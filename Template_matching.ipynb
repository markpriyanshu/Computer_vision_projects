```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
full = cv2.imread('C:/users/markp/Documents/computer vision/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/sammy.jpg')
full = cv2.cvtColor(full , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face = cv2.imread('C:/users/markp/Documents/computer vision/Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/sammy_face.jpg')
face = cv2.cvtColor(face , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(face)
```

```python
full_copy = full.copy()
```

```python
result = cv2.matchTemplate(full_copy, face, cv2.TM_CCOEFF )
```

```python
plt.imshow(result)
```

```python
cv2.minMaxLoc(result)
```

```python
height, width,channels = face.shape
```

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    full_img_copy = full.copy()
    method = eval(m)
    result = cv2.matchTemplate(full_img_copy, face, method)
    min_value,max_value,min_loc,max_loc = cv2.minMaxLoc(result)
    
    if m == 'cv2.TM_SQDIFF' or m == 'cv2.TM_SQDIFF_NORMED':
        top_left = min_loc
    else:
        top_left = max_loc
        
    bottom_right = (top_left[0]+width , top_left[1]+height)    
        
    cv2.rectangle(full_img_copy, top_left, bottom_right,(255,0,0), 10)    
    
    plt.subplot(121)
    plt.imshow(result)
    plt.title('Result of template matching')
    
    plt.subplot(122)
    plt.imshow(full_img_copy)
    plt.title('Detected point')
    plt.suptitle(m)
    
    plt.show()
    print()
    print()
        
```
